```python
import math
import re

# ---------------- NUMBER WORDS ----------------

number_words = {
    "zero":0,"one":1,"two":2,"three":3,"four":4,
    "five":5,"six":6,"seven":7,"eight":8,"nine":9,
    "ten":10,"eleven":11,"twelve":12,"thirteen":13,
    "fourteen":14,"fifteen":15,"sixteen":16,
    "seventeen":17,"eighteen":18,"nineteen":19,
    "twenty":20,"thirty":30,"forty":40,
    "fifty":50,"sixty":60,"seventy":70,
    "eighty":80,"ninety":90
}


def convert_number_words(text):

    for word, num in number_words.items():
        text = text.replace(word, str(num))

    return text


# ---------------- TEXT PREPROCESS ----------------

def preprocess_text(text):

    text = text.lower()

    text = convert_number_words(text)

    replacements = {
        "plus":"+",
        "add":"+",
        "minus":"-",
        "subtract":"-",
        "times":"*",
        "multiply":"*",
        "divide":"/",
        "power":"**"
    }

    for word, symbol in replacements.items():
        text = text.replace(word, symbol)

    return text


# ---------------- SCIENTIFIC FUNCTIONS ----------------

def convert_scientific(text):

    text = re.sub(r"sin (\d+)", r"math.sin(math.radians(\1))", text)
    text = re.sub(r"sine (\d+)", r"math.sin(math.radians(\1))", text)

    text = re.sub(r"cos (\d+)", r"math.cos(math.radians(\1))", text)
    text = re.sub(r"cosine (\d+)", r"math.cos(math.radians(\1))", text)

    text = re.sub(r"tan (\d+)", r"math.tan(math.radians(\1))", text)

    text = re.sub(r"log (\d+)", r"math.log10(\1)", text)

    text = re.sub(r"square root (\d+)", r"math.sqrt(\1)", text)
    text = re.sub(r"sqrt (\d+)", r"math.sqrt(\1)", text)

    text = re.sub(r"factorial (\d+)", r"math.factorial(\1)", text)

    return text


# ---------------- CALCULATE ----------------

def calculate_expression(expression):

    try:
        result = eval(expression, {"math": math})
        return result
    except Exception as e:
        print("Error:", e)
        return "Invalid Expression"


# ---------------- MAIN PROGRAM ----------------

voice_input = input("Speak your expression: ")

expression = preprocess_text(voice_input)

expression = convert_scientific(expression)

print("Converted Expression:", expression)

result = calculate_expression(expression)

print("Result:", result)
```
