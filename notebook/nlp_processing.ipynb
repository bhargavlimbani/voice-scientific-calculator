```python
import math
import re

def preprocess_text(text):
    text = text.lower()
    
    # Replace basic operators
    replacements = {
        "plus": "+",
        "minus": "-",
        "multiply by": "*",
        "multiplied by": "*",
        "times": "*",
        "divide by": "/",
        "divided by": "/",
        "power": "**",
    }
    
    for word, symbol in replacements.items():
        text = text.replace(word, symbol)
    
    return text


def convert_scientific_functions(text):
    
    # Sine
    text = re.sub(r"sine (\d+)", r"math.sin(math.radians(\1))", text)
    
    # Cosine
    text = re.sub(r"cosine (\d+)", r"math.cos(math.radians(\1))", text)
    
    # Tangent
    text = re.sub(r"tangent (\d+)", r"math.tan(math.radians(\1))", text)
    
    # Log base 10
    text = re.sub(r"log (\d+)", r"math.log10(\1)", text)
    
    # Natural log
    text = re.sub(r"ln (\d+)", r"math.log(\1)", text)
    
    # Square root
    text = re.sub(r"square root (\d+)", r"math.sqrt(\1)", text)
    
    # Factorial
    text = re.sub(r"factorial (\d+)", r"math.factorial(\1)", text)
    
    return text


def convert_text_to_expression(text):
    text = preprocess_text(text)
    text = convert_scientific_functions(text)
    return text


def calculate_expression(expression):
    try:
        result = eval(expression)
        return result
    except Exception as e:
        return "Invalid Expression"


# 🔥 TEST CASES
test_sentences = [
    "5 plus 3",
    "sine 30",
    "cosine 60 minus 1",
    "log 100",
    "square root 16",
    "factorial 5",
    "10 power 3"
]

for sentence in test_sentences:
    expression = convert_text_to_expression(sentence)
    result = calculate_expression(expression)
    print("Voice Input:", sentence)
    print("Converted:", expression)
    print("Result:", result)
    print("------")
```
